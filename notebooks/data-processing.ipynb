```python
# binary encoding for columns with only 2 unique values

df["gender"] = df["gender"].map({"Female": 0, "Male": 1})
df["Partner"] = df["Partner"].map({"No": 0, "Yes": 1})
df["Dependents"] = df["Dependents"].map({"No": 0, "Yes": 1})
df["PhoneService"] = df["PhoneService"].map({"No": 0, "Yes": 1})
df["PaperlessBilling"]= df["PaperlessBilling"].map({"No": 0, "Yes": 1})
df["Churn"] = df["Churn"].map({"No": 0, "Yes": 1})
```

```python
# one hot encoding for columns with more than 2 unique values
df = pd.get_dummies(df, columns=["Contract", "InternetService", "PaymentMethod"], drop_first=True)
```

```python
# handling "No phone service" & "No internet service" --> making then to No 

cols_to_clean = ["MultipleLines", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]
for col in cols_to_clean:
    df[col] = df[col].replace({"No phone service": "No", "No internet service": "No"})
    df[col] = df[col].map({"No": 0, "Yes": 1})
```

```python
df[["Contract_One year", "Contract_Two year",
    "InternetService_Fiber optic", "InternetService_No",
    "PaymentMethod_Credit card (automatic)", "PaymentMethod_Electronic check",
    "PaymentMethod_Mailed check"]] = df[["Contract_One year", "Contract_Two year",
                                         "InternetService_Fiber optic", "InternetService_No",
                                         "PaymentMethod_Credit card (automatic)", "PaymentMethod_Electronic check",
                                         "PaymentMethod_Mailed check"]].astype(int)
```
